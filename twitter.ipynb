```python

source=[]
language=[]
place=[]
rt=[]
for i in tweets:
    source.append(i.source)
    language.append(i.lang)
    place.append(i.place)
    rt.append(i.retweet_count)
```

```python
from tqdm import tqdm
source=[]
language=[]
place=[]
rt=[]
time=[]
for i in tqdm(range(1,12)):
    tweets=api.user_timeline("ShashiTharoor",page=i,count=200)
    
    for i in tweets:
        source.append(i.source)
        language.append(i.lang)
        place.append(i.place)
        rt.append(i.retweet_count)
        time.append(i.created_at)
```

```python
table=pd.DataFrame({'Source':source,'Language':language,'Place':place,'Rt':rt,'Time':time})
```
