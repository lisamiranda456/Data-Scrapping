### Developing the full logic

```python
url1='https://www.oyorooms.com/hotels-in-coimbatore/?page=4'
browser.get(url1)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
listings=soup.find_all('div',class_='hotelCardListing__descriptionWrapper')
```

```python
n=[]
add=[]
rating=[]
price=[]
for i in listings:
    name=i.find_all('h3',class_='listingHotelDescription__hotelName d-textEllipsis')
    if len(name)!=0:
        n.append(name[0].text)
    else:
        n.append(np.nan)
    addr=i.find_all('span',attrs={'itemprop':'streetAddress'})
    if len(addr)!=0:
        add.append(addr[0].text)
    else:
        add.append(np.nan)
    r=i.find_all('meta',attrs={'itemprop':'ratingValue'})
    if len(r)!=0:
        rating.append(r[0]['content'])
    else:
        rating.append(np.nan)
    p=i.find_all('span',class_='listingPrice__finalPrice')
    if len(p)!=0:
        price.append(p[0].text)
    else:
        price.append(np.nan)
    
```

```python
def my_data(x):
    browser.get(x)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    listings=soup.find_all('div',class_='hotelCardListing__descriptionWrapper')
    n=[]
    add=[]
    rating=[]
    price=[]
    for i in listings:
        name=i.find_all('h3',class_='listingHotelDescription__hotelName d-textEllipsis')
        if len(name)!=0:
            n.append(name[0].text)
        else:
            n.append(np.nan)
        addr=i.find_all('span',attrs={'itemprop':'streetAddress'})
        if len(addr)!=0:
            add.append(addr[0].text)
        else:
            add.append(np.nan)
        r=i.find_all('meta',attrs={'itemprop':'ratingValue'})
        if len(r)!=0:
            rating.append(r[0]['content'])
        else:
            rating.append(np.nan)
        p=i.find_all('span',class_='listingPrice__finalPrice')
        if len(p)!=0:
            price.append(p[0].text)
        else:
            price.append(np.nan)
    table=pd.DataFrame({'Name':n,'Address':add,'Ratings':rating,'Final_Price':price})
    return table
    
```

```python
url='https://www.oyorooms.com/hotels-in-coimbatore/?page=5'
my_data(url)
```

```python
base_url='https://www.oyorooms.com/hotels-in-coimbatore/?page={}'
```

```python
master={}
from tqdm import tqdm
for i in tqdm(range(1,6)):
    url=base_url.format(i)
    master[i]=my_data(url)
```

```python
bigtable=pd.concat([master[i] for i in master],axis=0)
```
