```python
def my_func(x):
    
    
    browser.get(x)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    tables=soup.find_all("table",class_="wikitable")
    rows=tables[0].find_all('tr')
    type1=[]
    date=[]
    dead=[]
    injured=[]
    location=[]
    details=[]
    perpetrator=[]
    part_of=[]

    for i in rows[1:]:
        type1.append(i.find_all("td")[1].text.strip())
        date.append(i.find_all("td")[0].text.strip("\n"))
        dead.append(i.find_all("td")[2].text.strip("\n"))
        injured.append(i.find_all("td")[3].text.strip("\n"))
        location.append(i.find_all("td")[4].text.strip("\n"))   
        details.append(i.find_all("td")[5].text.strip("\n"))
        perpetrator.append(i.find_all("td")[6].text.strip("\n"))
        part_of.append(i.find_all("td")[7].text.strip("\n")) 
    table=pd.DataFrame({'Type':type1,'Date':date,'Dead':dead,'Injured':injured,'Location':location,'Details':details,'Perpetrator':perpetrator,'Part':part_of})
    return table
```

```python
def my_func1(x):
    
    
    browser.get(x)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    tables=soup.find_all("table",class_="wikitable")
    rows=tables[0].find_all('tr')
    type1=[]
    date=[]
    dead=[]
    injured=[]
    location=[]
    details=[]
    perpetrator=[]
    part_of=[]
    url=[]

    for i in rows[1:]:
        type1.append(i.find_all("td")[1].text.strip("\n"))
        date.append(i.find_all("td")[0].text.strip("\n"))
        dead.append(i.find_all("td")[2].text.strip("\n"))
        injured.append(i.find_all("td")[3].text.strip("\n"))
        location.append(i.find_all("td")[4].text.strip("\n"))   
        details.append(i.find_all("td")[5].text.strip("\n"))
        url.append(x)
        perpetrator.append(i.find_all("td")[6].text.strip("\n"))
        try:
            part_of.append(i.find_all("td")[7].text.strip("\n")) 
        except:
            part_of.append("data not available")
    table=pd.DataFrame({'Type':type1,'Date':date,'Dead':dead,'Injured':injured,'Location':location,'Details':details,'Perpetrator':perpetrator,'Part':part_of,"Url":url})
    return table
```
